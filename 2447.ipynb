```python
def star(size, x, y):
    if size == 1:
        S[y][x] = '*'
    else:
        next_size = size // 3
        for dy in range(0, 3):
            for dx in range(0,3):
                if dy != 1 or dx != 1:
                    star(next_size, x+dx*next_size, y+dy*next_size)

N = int(input())
S = [[" " for i in range(0,N)] for i in range(0,N)]
star(N,0,0)

for i in range(0, N):
    for j in range(0, N):
        print(S[i][j], end='')
    print()
```
